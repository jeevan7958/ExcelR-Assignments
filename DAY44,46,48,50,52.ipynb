```python
#Day 44
def chatbot():
    while True:
        user_input = input("You: ").strip()
        if user_input.lower() == "hello":
            print("Chatbot: Hi! How can I help you?")
        elif user_input.lower() == "bye":
            print("Chatbot: Goodbye! Have a nice day!")
            break
        else:
            print("Chatbot: I'm sorry, I don't understand.")

# Run the chatbot
chatbot()
```

```python
#DAY 46
def chatbot():
    responses = {
        "Hello": "Hi there! How can I help you?",
        "How are you?": "I'm just a bot, but I'm doing great! How about you?",
        "Bye": "Goodbye! Have a great day!"
    }

    while True:
        user_input = input("You: ")
        if user_input in responses:
            print("Bot:", responses[user_input])
        else:
            print("Bot: I'm not sure how to respond to that.")

        if user_input == "Bye":
            break

if __name__ == "__main__":
    chatbot()
```

```python
#Day 48
from transformers import pipeline

def summarize_text(text):
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    summary = summarizer(text, max_length=50, min_length=10, do_sample=False)
    return summary[0]['summary_text']

if __name__ == "__main__":
    text = ("Artificial Intelligence (AI) is a rapidly advancing field that is transforming "
            "various industries, including healthcare, finance, and transportation. "
            "It enables machines to learn from data and make decisions with minimal human intervention.")

    summary = summarize_text(text)
    print("Summary:", summary)
```

```python
#Day 50
import numpy as np

def attention_scores():
    np.random.seed(42)  # For reproducibility

    # Generate random query, key, and value vectors
    query = np.random.rand(1, 3)  # 1x3 vector
    key = np.random.rand(3, 3)    # 3x3 matrix
    value = np.random.rand(3, 3)  # 3x3 matrix

    # Compute dot product between query and key (scaled by sqrt of key dimension)
    scores = np.dot(query, key.T) / np.sqrt(key.shape[1])

    # Apply softmax to obtain attention weights
    attention_weights = np.exp(scores) / np.sum(np.exp(scores), axis=1, keepdims=True)

    # Compute final attention output
    attention_output = np.dot(attention_weights, value)

    return attention_output

if __name__ == "__main__":
    result = attention_scores()
    print("Attention Output:", result)
```

```python
#Day 52
from transformers import BertTokenizer

def tokenize_sentence(sentence):
    # Load the pre-trained BERT tokenizer
    tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

    # Tokenize the sentence
    tokens = tokenizer.tokenize(sentence)
    token_ids = tokenizer.encode(sentence, add_special_tokens=True)

    return tokens, token_ids

if __name__ == "__main__":
    sentence = "Hugging Face makes NLP easy!"
    tokens, token_ids = tokenize_sentence(sentence)
    print("Tokens:", tokens)
    print("Token IDs:", token_ids)
```
